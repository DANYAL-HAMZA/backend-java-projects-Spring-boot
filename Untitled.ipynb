```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
try:
    data = pd.read_csv('Koforidua_rr.csv')
    print("Dataset loaded successfully. Columns:", data.columns)
except FileNotFoundError:
    raise FileNotFoundError("Dataset 'Ejura_rr.csv' not found in the working directory.")

# Validate required columns
required_columns = ['Year', 'Month', 'Days', 'Rainfall']
missing_columns = set(required_columns) - set(data.columns)
if missing_columns:
    raise ValueError(f"Dataset is missing required columns: {missing_columns}")

# Convert 'Year', 'Month', 'Days' columns to numeric
data['Year'] = pd.to_numeric(data['Year'], errors='coerce')
data['Month'] = pd.to_numeric(data['Month'], errors='coerce')
data['Days'] = pd.to_numeric(data['Days'], errors='coerce')

# Filter valid combinations of month and day
data = data[(data['Month'].between(1, 12)) & (data['Days'].between(1, 31))]

# Create a new 'Date' column and handle invalid dates
data['Date'] = pd.to_datetime(data[['Year', 'Month', 'Days']], errors='coerce')
invalid_dates = data[data['Date'].isna()]
if not invalid_dates.empty:
    print(f"Warning: {len(invalid_dates)} rows with invalid dates removed.")
data = data.dropna(subset=['Date'])

# Filter data for the years 1980 to 2020
data = data[(data['Date'].dt.year >= 1980) & (data['Date'].dt.year <= 2020)]

# Convert 'Rainfall' column to numeric and handle missing values
data['Rainfall'] = pd.to_numeric(data['Rainfall'], errors='coerce')

# Drop rows where 'Rainfall' is NaN
if data['Rainfall'].isna().any():
    print(f"Warning: {data['Rainfall'].isna().sum()} rows with invalid 'Rainfall' values removed.")
data = data.dropna(subset=['Rainfall'])

# Set 'Date' as the index for resampling
data.set_index('Date', inplace=True)

# Calculate monthly total rainfall (use 'MS' for Month Start)
monthly_rainfall = data.resample('MS').sum()

# Create a 'Decade' column for grouping
monthly_rainfall['Decade'] = (monthly_rainfall.index.year // 10) * 10

# Group by month and decade, then sum rainfall
decadal_rainfall = monthly_rainfall.groupby([monthly_rainfall.index.month, 'Decade'])['Rainfall'].sum().unstack()

# Filter decades to ensure they end at 2010-2020
decadal_rainfall = decadal_rainfall.loc[:, [1980, 1990, 2000, 2010]]

# Rename columns to match the format in the uploaded image
decadal_rainfall.columns = [
    f"{decade + 1}-{decade + 10}" for decade in decadal_rainfall.columns
]

# Plotting decadal monthly rainfall
plt.figure(figsize=(12, 6))
decadal_rainfall.plot(kind='bar', width=0.8, ax=plt.gca())
plt.title('Figure 4-1a. Total Decadal Monthly Rainfall')
plt.xlabel('Months')
plt.ylabel('Rainfall (mm)')
plt.xticks(
    ticks=range(12),
    labels=['JAN', 'FEB', 'MAR', 'APR', 'MAY', 'JUN', 'JUL', 'AUG', 'SEP', 'OCT', 'NOV', 'DEC'],
    rotation=45
)
plt.legend(title='Decade')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# Summary statistics
summary_stats = data['Rainfall'].describe()
print("Summary Statistics:\n", summary_stats)

# Count the number of days with rain
days_with_rain = data[data['Rainfall'] > 0].shape[0]
print(f"Total days with rain: {days_with_rain}")

# Maximum daily rainfall
max_rainfall = data['Rainfall'].max()
max_rainfall_date = data['Rainfall'].idxmax()
print(f"Maximum daily rainfall: {max_rainfall} mm on {max_rainfall_date.date()}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
try:
    data = pd.read_csv('Koforidua_rr.csv')
    print("Dataset loaded successfully.")
except FileNotFoundError:
    raise FileNotFoundError("Dataset 'Ejura_rr.csv' not found in the working directory.")

# Ensure 'Rainfall' is numeric
data['Rainfall'] = pd.to_numeric(data['Rainfall'], errors='coerce')

# Convert 'Year', 'Month', 'Days' columns to numeric
data['Year'] = pd.to_numeric(data['Year'], errors='coerce')
data['Month'] = pd.to_numeric(data['Month'], errors='coerce')
data['Days'] = pd.to_numeric(data['Days'], errors='coerce')

# Create a new 'Date' column and handle invalid dates
data['Date'] = pd.to_datetime(data[['Year', 'Month', 'Days']], errors='coerce')
data = data.dropna(subset=['Date'])

# Filter data for the years 1980 to 2020
data = data[(data['Date'].dt.year >= 1981) & (data['Date'].dt.year <= 2020)]

# Define the rainfall categorization function
def categorize_rainfall(rainfall):
    if pd.isna(rainfall):
        return None  # Handle missing data
    elif 0 < rainfall <= 0.9:
        return 'Very Light'
    elif 1.0 <= rainfall <= 10.0:
        return 'Light'
    elif 11.0 <= rainfall <= 30.0:
        return 'Moderate'
    elif 31.0 <= rainfall <= 70.0:
        return 'Heavy'
    elif 71.0 <= rainfall <= 150.0:
        return 'Very Heavy'
    elif rainfall > 150.0:
        return 'Extremely Heavy'
    else:
        return None  # For 0 mm or other cases

# Apply the rainfall classification
data['Rainfall Class'] = data['Rainfall'].apply(categorize_rainfall)

# Filter relevant columns and ensure no missing data
data = data.dropna(subset=['Rainfall', 'Rainfall Class'])
data['Month'] = data['Date'].dt.month

# Debug: Check distribution of rainfall classes
print("Rainfall Class Distribution:")
print(data['Rainfall Class'].value_counts())

# Function to plot rainfall data by year
def plot_rainfall_by_year(data):
    unique_years = sorted(data['Year'].unique())
    num_years = len(unique_years)

    # Create subplots grid
    rows = num_years // 2 + (num_years % 2 > 0)
    fig, axes = plt.subplots(nrows=rows, ncols=2, figsize=(20, rows * 6))
    axes = axes.flatten()

    for i, year in enumerate(unique_years):
        yearly_data = data[data['Year'] == year]
        monthly_rainfall = yearly_data.groupby(['Month', 'Rainfall Class']).size().unstack(fill_value=0)

        # Ensure all rainfall classes are included
        for category in ['Very Light', 'Light', 'Moderate', 'Heavy', 'Very Heavy', 'Extremely Heavy']:
            if category not in monthly_rainfall.columns:
                monthly_rainfall[category] = 0

        # Sort columns
        monthly_rainfall = monthly_rainfall[['Very Light', 'Light', 'Moderate', 'Heavy', 'Very Heavy', 'Extremely Heavy']]

        # Plot stacked bar for the year
        monthly_rainfall.plot(
            kind='bar', 
            stacked=True, 
            ax=axes[i], 
            width=0.8, 
            legend=False, 
            color=['#a6cee3', '#1f78b4', '#b2df8a', '#33a02c', '#fb9a99', '#e31a1c']
        )
        axes[i].set_title(f"Year: {year}", fontsize=12)
        axes[i].set_xlabel('Months')
        axes[i].set_ylabel('Rainfall (mm)')
        axes[i].tick_params(axis='x', labelrotation=0)
        axes[i].grid(axis='y', linestyle='--', alpha=0.7)

    # Remove unused subplots
    for j in range(i + 1, len(axes)):
        fig.delaxes(axes[j])

    # Add a legend outside the plots
    fig.legend(
        ['Very Light', 'Light', 'Moderate', 'Heavy', 'Very Heavy', 'Extremely Heavy'], 
        loc='upper center', 
        ncol=6, 
        fontsize=12, 
        bbox_to_anchor=(0.5, 0.97)
    )

    plt.tight_layout(rect=[0, 0, 1, 0.95])
    plt.suptitle('Monthly Rainfall Class Contributions by Year (1980-2020)', fontsize=18)
    plt.show()

# Plot the data
plot_rainfall_by_year(data)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the rainfall data from the CSV file
data = pd.read_csv('Koforidua_rr.csv')

# Convert 'Rainfall' column to numeric, forcing errors to NaN
data['Rainfall'] = pd.to_numeric(data['Rainfall'], errors='coerce')

# Filter data from 1981 to 2020
data = data[(data['Year'] >= 1981) & (data['Year'] <= 2020)]

# Group by Year and Month to calculate the monthly total rainfall
monthly_rainfall = data.groupby(['Year', 'Month'])['Rainfall'].sum().reset_index()

# Calculate the mean and standard deviation for each month over the years
monthly_mean = monthly_rainfall.groupby('Month')['Rainfall'].mean()
monthly_std = monthly_rainfall.groupby('Month')['Rainfall'].std()

# Merge means and std deviations back into the monthly data
monthly_rainfall = monthly_rainfall.merge(monthly_mean.rename('Mean'), on='Month')
monthly_rainfall = monthly_rainfall.merge(monthly_std.rename('StdDev'), on='Month')

# Calculate the Standardized Rainfall Anomaly Index (SRAI)
monthly_rainfall['SRAI'] = (monthly_rainfall['Rainfall'] - monthly_rainfall['Mean']) / monthly_rainfall['StdDev']

# Calculate the yearly average SRAI
yearly_srai = monthly_rainfall.groupby('Year')['SRAI'].mean().reset_index()

# Create a stem plot for SRAI
plt.figure(figsize=(12, 6))
plt.stem(yearly_srai['Year'], yearly_srai['SRAI'], basefmt=" ")

# Customize the plot
plt.title('Ejura Standard Anomaly Rainfall (mm) by Year (1981-2020)')
plt.xlabel('Year')
plt.ylabel('Standard Anomaly Rainfall (mm)')
plt.axhline(0, color='gray', linestyle='--')  # Reference line at 0
plt.grid()

# Display the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV data
df = pd.read_csv('Koforidua_rr.csv')

# Define rainfall classification function
def classify_rainfall(rainfall):
    if pd.isna(rainfall) or rainfall == 0:
        return None  # Exclude "No Rain"
    elif rainfall < 2.5:
        return 'Very Light'
    elif rainfall < 7.6:
        return 'Light'
    elif rainfall < 20:
        return 'Moderate'
    elif rainfall < 50:
        return 'Heavy'
    elif rainfall < 100:
        return 'Very Heavy'
    else:
        return 'Extremely Heavy'

# Apply classification function to the Rainfall column
df['Rainfall Class'] = df['Rainfall'].apply(classify_rainfall)

# Filter out rows with no rain
df_filtered = df.dropna(subset=['Rainfall Class'])

# Count occurrences of each rainfall class
rainfall_counts = df_filtered['Rainfall Class'].value_counts()

# Create a pie chart
plt.figure(figsize=(8, 6))
plt.pie(rainfall_counts, labels=rainfall_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Rainfall Classification Distribution')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import linregress

# Load the rainfall data from the CSV file
data = pd.read_csv('Koforidua_rr.csv')

# Convert 'Rainfall' column to numeric, forcing errors to NaN
data['Rainfall'] = pd.to_numeric(data['Rainfall'], errors='coerce')

# Filter data from 1980 to 2020
data = data[(data['Year'] >= 1980) & (data['Year'] <= 2020)]

# Group by Year and Month to calculate the monthly total rainfall
monthly_rainfall = data.groupby(['Year', 'Month'])['Rainfall'].sum().reset_index()

# Calculate the mean and standard deviation for each month over the years
monthly_mean = monthly_rainfall.groupby('Month')['Rainfall'].mean()
monthly_std = monthly_rainfall.groupby('Month')['Rainfall'].std()

# Merge means and std deviations back into the monthly data
monthly_rainfall = monthly_rainfall.merge(monthly_mean.rename('Mean'), on='Month')
monthly_rainfall = monthly_rainfall.merge(monthly_std.rename('StdDev'), on='Month')

# Calculate the Standardized Rainfall Anomaly Index (SRAI)
monthly_rainfall['SRAI'] = (monthly_rainfall['Rainfall'] - monthly_rainfall['Mean']) / monthly_rainfall['StdDev']

# Descriptive Statistics
print("Descriptive Statistics of SRAI:")
print(monthly_rainfall['SRAI'].describe())

# Visualization of SRAI over time using a stem plot
plt.figure(figsize=(12, 6))
plt.stem(monthly_rainfall['Year'], monthly_rainfall['SRAI'], basefmt=" ")
plt.title('SRAI Over Time (1980-2020)')
plt.xlabel('Year')
plt.ylabel('SRAI')
plt.axhline(0, color='red', linestyle='--', label='Average SRAI (0)')
plt.legend()
plt.grid()
plt.show()

# Trend Analysis
slope, intercept, r_value, p_value, std_err = linregress(monthly_rainfall['Year'], monthly_rainfall['SRAI'])
print(f"Trend Slope: {slope}, R-squared: {r_value**2}")

# Drought and Flood Analysis
threshold_high = 2  # Example threshold for high anomalies
threshold_low = -2   # Example threshold for low anomalies

extreme_highs = monthly_rainfall[monthly_rainfall['SRAI'] > threshold_high]
extreme_lows = monthly_rainfall[monthly_rainfall['SRAI'] < threshold_low]

print("\nExtreme High SRAI Events:")
print(extreme_highs[['Year', 'Month', 'Rainfall', 'SRAI']])

print("\nExtreme Low SRAI Events:")
print(extreme_lows[['Year', 'Month', 'Rainfall', 'SRAI']])

# Further Analysis: Visualizing Extreme Events
plt.figure(figsize=(12, 6))
sns.scatterplot(data=monthly_rainfall, x='Year', y='SRAI', color='blue', label='SRAI')
sns.scatterplot(data=extreme_highs, x='Year', y='SRAI', color='red', label='Extreme Highs')
sns.scatterplot(data=extreme_lows, x='Year', y='SRAI', color='green', label='Extreme Lows')
plt.title('Extreme SRAI Events (1980-2020)')
plt.xlabel('Year')
plt.ylabel('SRAI')
plt.axhline(0, color='black', linestyle='--')
plt.legend()
plt.grid()
plt.show()
```
